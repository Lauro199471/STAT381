```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def draw_z_score(x, cond, mu, sigma, title,alpha,z0):
    y = norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    #plt.fill_between(z, 0, norm.pdf(z, mu, sigma))
    plt.title(title)
    plt.axvline(x=z0, color='brown', linestyle='--')
    plt.axvline(x=alpha, color='blue', linestyle='--')
    plt.show()
```

```python
x = np.arange(0,.02,0.000001)

z0 = .0009
alpha = .01

draw_z_score(x, x<z0, 0, 1, '',alpha,z0)
```

```python
x = np.arange(-5,5,0.00001)

z0 = -1.645
alpha = -0.56

draw_z_score(x, x>z0, 0, 1, 'z<-1.645',alpha,z0)
```

```python
def draw_z_score2(x, cond1,cond2, mu, sigma, title,alpha,z0):
    y = norm.pdf(x, mu, sigma)
    z1 = x[cond1]
    z2 = x[cond2]
    plt.plot(x, y)
    plt.fill_between(z1, 0, norm.pdf(z1, mu, sigma))
    plt.fill_between(0, z2, norm.pdf(z2, mu, sigma))
    plt.title(title)
    plt.axvline(x=z0, color='brown', linestyle='--')
    plt.axvline(x=alpha, color='blue', linestyle='--')
    plt.show()
```

```python
x = np.arange(-3,3,0.001)
z0 = 2.576
draw_z_score2(x, (x<z0) , (x>z0), 0, 1, '-0.75<z<0.75',alpha,z0)
```
